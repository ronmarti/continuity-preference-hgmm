```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file1 = '../.out/illustrations/vanilla_kr210.csv'
file2 = '../.out/illustrations/cont_kr210.csv'
# file1 = '../.out/illustrations/vanilla_kr5.csv'
# file2 = '../.out/illustrations/cont_kr5.csv'
df_van = pd.read_csv(file1)
df_cont = pd.read_csv(file2)
```

```python
vanilla_series = df_van['0']
cont_series = df_cont['0']
merged_data = {
    'Plain HGMM': vanilla_series,
    'Continuity preference HGMM': cont_series
}
df = pd.DataFrame(merged_data) 
fig, axes = plt.subplots(1, figsize=(7, 2.5))
axes.plot(df['Plain HGMM'], label='Without continuity preference', LineWidth=2)
axes.plot(df['Continuity preference HGMM'], label='With continuity preference', LineWidth=2)
axes.set_xlim([0, 301])
axes.legend(loc='lower right')
axes.set_ylabel('$a_{11}$ [-]')
axes.set_xlabel('Sample $t$')
plt.tight_layout()
```

```python
out_path = '../.out/illustrations/continuity_illustration_kr210.pdf'
fig.savefig(out_path)
```
