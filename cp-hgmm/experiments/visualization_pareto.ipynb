```python
import numpy as np
import pandas as pd

from pathlib import Path

import browse_dlg

import matplotlib.pyplot as plt
import mplcursors

import seaborn as sns

# from matplotlib import rc
# import matplotlib.pylab as plt

# rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
# rc('text', usetex=True)

%matplotlib widget
```

```python
def load_preceding_experiments(root='.'):
    dfs_list = []
    preceding_experiments_collected = []
    while restore_filepath := browse_dlg.gui_fname(root):
        preceding_experiments_collected.append(restore_filepath.decode())
    for filepath in preceding_experiments_collected:
        df = pd.read_csv(Path(filepath), index_col='trial_index')
        dfs_list.append(df)
        
    print(preceding_experiments_collected)
    total_df = pd.concat(dfs_list, ignore_index=True)
    return total_df
```

# Load Experiment #

```python
exp_root = Path(r'..\.out\experiments')
exp_data_df = load_preceding_experiments(exp_root)
exp_data_df
```

```python
exp_data_df['kappa'] = exp_data_df['kappa_sqrt']**2
# exp_data_df['kappa_sqrt'] = np.sqrt(exp_data_df['kappa'])
# exp_data_df.sort_values('num_states')
```

```python
from matplotlib import colors
my_cmap = sns.color_palette("rocket", as_cmap=True)
# my_cmap = sns.color_palette("coolwarm", as_cmap=True)
# my_cmap = plt.get_cmap('hot')
# sns.scatterplot(
#     data=exp_data_df, x="gap_elpd", y="gap_pos_tr_val_elpd", size=2, c=exp_data_df['kappa_sqrt'], cmap=my_cmap,
#     legend="full"
# )
# sns.set_context('paper')
# sns.set_style('whitegrid')

sc = exp_data_df.plot.scatter('gap_elpd', 'gap_pos_tr_val_elpd', s=30, c='kappa_sqrt', cmap=my_cmap, norm=colors.PowerNorm(gamma=0.25))
sc.set_xlabel('$g_{elpd}$')
sc.set_ylabel('$g_{var}$')
fig = plt.gcf()

cax = fig.get_axes()[1]
cax.set_ylabel('$\sqrt{\kappa}$')

# we can change it by setting hover to True
cursor = mplcursors.cursor(sc, hover=True)# by default the annotation displays the xy positions
# this is to change it to the countries name
@cursor.connect("add")
def on_add(sel):
    arm_name = exp_data_df.loc[sel.target.index, 'arm_name']
    kappa = exp_data_df.loc[sel.target.index, 'kappa']
    kappa_sqrt = exp_data_df.loc[sel.target.index, 'kappa_sqrt']
    num_states = exp_data_df.loc[sel.target.index, 'num_states']
    text = f'Arm: {arm_name}\nKappa: {kappa}\nKappa_sqrt: {kappa_sqrt}\nNum_states: {num_states}'
    sel.annotation.set(text=text)
```
